# Классификация

### Метод ближайших соседей

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
X = df.drop(columns=['ID', 'ZIP Code', 'Personal Loan'])
y = df['Personal Loan']
```

```python
split_ratio = 0.7
split_index = int(len(X) * split_ratio)
X_train, X_test = X[:split_index], X[split_index:]
y_train, y_test = y[:split_index], y[split_index:]
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
```

```python
y_test_predict = knn.predict(X_test)
```

```python
knn_pred_test = knn.predict(X_test)
```

```python
print("knn_pred_test[:20]:", knn_pred_test[:20])
print("y_test[:20]:", y_test[:20])
```

```python
y_train_predict = knn.predict(X_train)
y_test_predict = knn.predict(X_test)

train_accuracy = np.mean(y_train_predict == y_train)
test_accuracy = np.mean(y_test_predict == y_test)

print("Тренировачна часть:", train_accuracy)
print("Тестовая часть:", test_accuracy)
```
